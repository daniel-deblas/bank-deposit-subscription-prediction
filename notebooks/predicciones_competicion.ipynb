```python
import pandas as pd
import numpy as np
import joblib

# Cargarmos el modelo final entrenado
modelo = joblib.load('modelo_final.joblib')

# Cargamos los datos de la competición
df_comp = pd.read_pickle('bank_competition.pkl')

# Replicamos el preproceso manual de la variable 'pdays'
df_comp['previous_contact'] = np.where(df_comp['pdays'] == -1, 0, 1)
df_comp.loc[df_comp['pdays'] == -1, 'pdays'] = np.nan

# Predecimos y exportamos los resultados
# El modelo devuelve 0 y 1, los mapeamos a 'no' y 'yes' según lo esperado
predicciones_comp = modelo.predict(df_comp)
predicciones_etiquetas = np.where(predicciones_comp == 1, 'yes', 'no')

# Guardamos las predicciones en formato CSV
df_pred = pd.DataFrame({'deposit_prediction': predicciones_etiquetas})
df_pred.to_csv('predicciones.csv', index=False)
print("Predicciones guardadas exitosamente en 'predicciones.csv'")
```
